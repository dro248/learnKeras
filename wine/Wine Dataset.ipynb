### Preprocessing Data

````python
# Add 'type' column to 'red' with value 1
red['type'] = 1

# Add 'type' column to 'white' with value 0
white['type'] = 0

# Append white list to red
wines = red.append(white, ignore_index=True)
wines
````

````python
import seaborn as sns
corr = wines.corr()
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
# sns.plt.show()
````

#### Training & Test Sets

````python
# Import `train_test_split` from `sklearn.model_selection`
from sklearn.model_selection import train_test_split

# Specify the data 
X=wines.iloc[:,0:11]

# Specify the target labels and flatten the array 
y=np.ravel(wines.type)

# Split the data up in train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
````

####  Standardize the data

````python
# Import `StandardScaler` from `sklearn.preprocessing`
from sklearn.preprocessing import StandardScaler

# Define the scaler 
scaler = StandardScaler().fit(X_train)

# Scale the train set
X_train = scaler.transform(X_train)

# Scale the test set
X_test = scaler.transform(X_test)
````

## Question: Can we predict whether a wine is red or white by looking at its chemical properties?


####  0. Intro
A type of network that performs well on this type of problem is a multi-layer perceptron. This type of neural network is often fully connected, meaning that you're looking to build a fairly simple stack of fully-connected layers to solve this problem. As for the activation function, we will use 'relu'.

#### 1. Getting Started
How do you start building yoru multi-layer perceptron? A quick way to get started is to use the Keras Sequential model: it's a linear stack of layers. You can easily create the model by passing a list of layer instances to the constructor, which you set up by running model = Sequential().

#### 2. Structure
- input layer
- n hidden layers
- output layer

In our case, we will have to use a **Dense** layer, that is, a fully-connected layer. Dense layers implement the following operation:
```python
output = activation(dot(input, kernel) + bias)
```
> Note that without the activation function, your **Dense** layer would consist only ofo two linear operations:
> - A dot product
> - an addition


````python
# Import `Sequential` from `keras.models`
from keras.models import Sequential

# Import `Dense` from `keras.layers`
from keras.layers import Dense

# Initialize the constructor
model = Sequential()

# Add an input layer 
model.add(Dense(12, activation='relu', input_shape=(11,)))

# Add one hidden layer 
model.add(Dense(8, activation='relu'))

# Add an output layer 
model.add(Dense(1, activation='sigmoid'))
````

### Compile and Fit

In compiling the model, we configure the model with the *adam* optimizer and the *binary_crossentropy* loss function. We also monitor the accuracy during the training by passing ['accuracy'] to the metrics argument.

The **Optimizer** and the **loss** are two arguments that are required if you want to compile the model. 

#### Optimizers
Some of the most popular optimization algorithms used are:
- Stochastic Gradient Descent (SGD)
- ADAM
- RMSprop

#### Loss

The choice for a loss function depends on the task that you have at hand. 
1. For *Regression* problems: use **Mean Squared Error** (MSE).
2. For *Binary Classification* problems (our problem): use **binary_crossentropy**.
3. For *Multi-class Classification* problems: use **categorical_crossentropy**.

````python
# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy']) 

# train the model over 20 epochs
model.fit(X_train, y_train,epochs=20, batch_size=1, verbose=1)
````

### Predict Values

````python
y_pred = model.predict(X_test)
````

````python
# 5th Predicted value
y_pred[:5]
````

````python
# Actual Value of 5th item
y_test[:5]
````

###  Evaluate Model


##### Evaluate using evaluate method
Now that we have built our model with your training data, let's evaluate its performance. We can visually compare the predictions with the actual test labels (*y_test*), or we can use all types of metrics to determine the actual performace. In our case, we'll use the *evaluate()* method

````python
score = model.evaluate(X_test, y_test, verbose=1)
print(score)
````

#### Evaluate using Confusion Matrix

````python
# Import the modules from `sklearn.metrics`
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, cohen_kappa_score

# Confusion matrix
confusion_matrix(y_test, y_pred.round())
````

#### Precision Score
